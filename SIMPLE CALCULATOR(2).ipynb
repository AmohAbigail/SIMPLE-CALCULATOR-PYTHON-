```python
num1 = float(input("Enter first number: "))
operator = input("Enter operator (( +, -, *, /, //, %, **: ")
num2 = float(input("Enter second number: "))

if operator =="+":
    result = num1 + num2
elif operator =="-":
    result = num1 - num2
elif operator =="*":
    result = num1 * num2
elif operator =="/":
    if num2 == 0:
         print("Error: Division by zero is not allowed.")
    else:
         result = num1 / num2
         print(f"Result: {num1} / {num2} = {result}")
elif operator == "//":
  if num2 == 0:
    print("Error: Division by zero is not allowed.")
  else:
       result = num1 // num2
       print(f"Result: {num1} // {num2} = {result}")
elif operator == "%":
  if num2 == 0:
    print("Error: Division by zero is not allowed.")
  else:
       result = num1 % num2
       print(f"Result: {num1} % {num2} = {result}")
elif operator == "**":
   result = num1 ** num2
   print(f"Result: {num1} ** {num2} = {result}")
else:
     print("Invalid operator.")
if operator in ["+", "-", "*", "**"]:
    print(f"Result: {num1} {operator} {num2} = {result}")

print("Complete Assignment 2 by implementing the simple calculator.")    
```
